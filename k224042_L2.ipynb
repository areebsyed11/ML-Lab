```python
# Print total number of features and records
num_features = df.shape[1]
num_records = df.shape[0]
print(f"Total number of features: {num_features}")
print(f"Total number of records: {num_records}")

def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i] 
                col_corr.add(colname)
    
    return col_corr
corr_features = correlation(df, 0.65)
df_reduced = df.drop(corr_features, axis=1)

print(f"Number of features to drop: {len(corr_features)}")
print(f"Reduced DataFrame shape: {df_reduced.shape}")
```

```python
df['Gold'] = df['Gold'].fillna(df['Gold'].mean())
df['Silver'] = df['Silver'].fillna(df['Silver'].mean())
df['Bronze'] = df['Bronze'].fillna(df['Bronze'].mean())
df['Total'] = df['Total'].fillna(df['Total'].mean())

missing_data_after_imputation = df.isnull().sum()
missing_data_after_imputation
```

```python
counts = df['Total'].value_counts()
print(counts)

# Checking if the dataset is balanced
total_count = len(df)
for category, count in counts.items():
    proportion = count / total_count
    print(f"Category {category} proportion: {proportion:.3f}")
threshold = 0.4  
is_balanced = all(count / total_count > threshold for count in counts)
print("\nIs the dataset balanced?", is_balanced)
```

```python
#E
from sklearn.utils import resample

df_majority = df[df['Total'] == df['Total'].value_counts().idxmax()]
df_minority = df[df['Total'] == df['Total'].value_counts().idxmin()]
majority_class_size = len(df_majority)
minority_class_size = len(df_minority)

if majority_class_size > 0 and minority_class_size > 0:
    df_minority_upsampled = resample(df_minority, replace=True,n_samples=majority_class_size,random_state=42) 

    df_balanced = pd.concat([df_majority, df_minority_upsampled])

    df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

    print("New class distribution after Random Oversampling:")
    print(df_balanced['Total'].value_counts())
else:
    print("Error: Majority or minority class size is 0, cannot perform resampling.")
```
