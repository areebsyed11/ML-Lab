```python
# Task 2
import pandas as pd
import numpy as np

data = {
    "ID": range(1, 16),
    "AGE": ["Young", "Young", "Young", "Young", "Young", "Middle", "Middle", "Middle", "Middle", "Middle",
            "Old", "Old", "Old", "Old", "Old"],
    "JOB_STATUS": [False, False, True, True, False, False, False, True, True, False,
                   True, True, False, True, False],
    "OWNS_HOUSE": [False, False, False, True, False, False, False, True, True, False,
                   True, True, False, True, False],
    "CREDIT_RATING": ["Fair", "Good", "Good", "Fair", "Fair", "Good", "Good", "Good",
                      "Excellent", "Excellent", "Excellent", "Excellent", "Good", "Excellent", "Fair"],
    "CLASS": ["No", "No", "Yes", "Yes", "No", "No", "Yes", "Yes", "Yes", "Yes",
              "Yes", "Yes", "Yes", "Yes", "No"]
}

df = pd.DataFrame(data)

#Calculate entropy
def entropy(column):
    counts = column.value_counts()
    probabilities = counts / len(column)
    return -np.sum(probabilities * np.log2(probabilities))

# Calculate Information Gain
def information_gain(data, attribute, target):
    
    # Entropy of Entire dataset
    total_entropy = entropy(data[target])
    
    # Entropy for each attribute
    attr_entropy = 0
    for value in data[attribute].unique():
        subset = data[data[attribute] == value]
        attr_entropy += (len(subset) / len(data)) * entropy(subset[target])

    return total_entropy - attr_entropy

# IG of each attribute
attributes = ['AGE', 'JOB_STATUS', 'OWNS_HOUSE', 'CREDIT_RATING']
ig_results = {}

for attribute in attributes:
    ig_results[attribute] = information_gain(df, attribute, 'CLASS')

for attribute, ig in ig_results.items():
    print(f'Information Gain for {attribute}: {ig:.4f}')

# Highest Information Gain
root_node = max(ig_results, key=ig_results.get)
print("Root Node: ",root_node)
```

```python
# Task 4
import pandas as pd
import numpy as np

data = {
    "Student": range(1, 11),
    "Prior_Experience": ["Yes", "No", "Yes", "No", "Yes", "No", "No", "Yes", "Yes", "Yes"],
    "Course": ["Programming", "Programming", "History", "Programming", "English", "Programming", "Programming", "Mathematics", "Programming", "Programming"],
    "Time": ["Day", "Day", "Night", "Night", "Day", "Day", "Day", "Night", "Night", "Night"],
    "Liked": ["Yes", "No", "No", "Yes", "Yes", "No", "No", "Yes", "Yes", "No"]
}

df = pd.DataFrame(data)

# Calculate Gini Index
def gini_index(column):
    counts = column.value_counts()
    probabilities = counts / len(column)
    return 1 - np.sum(probabilities ** 2)

# Calculate Gini for a split
def gini_for_split(data, attribute, target):
    gini_split = 0
    for value in data[attribute].unique():
        subset = data[data[attribute] == value]
        gini_split += (len(subset) / len(data)) * gini_index(subset[target])
    return gini_split

# Gini Index for each attribute
attributes = ['Prior_Experience', 'Course', 'Time']
gini_results = {}

for attribute in attributes:
    gini_results[attribute] = gini_for_split(df, attribute, 'Liked')

for attribute, gini in gini_results.items():
    print(f'Gini Index for {attribute}: {gini:.4f}')

# Lowest Gini Index attribute 
root_node = min(gini_results, key=gini_results.get)
print("Root Node: ", root_node)
```
